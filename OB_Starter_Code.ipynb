```python
#data cleaning starts here. started with a much larger file for indiana obgyns sorted file but was causing problems on reruns so removed lines of code relevant to that process

IndianaData = "Indiana_OBGYNS_Final.csv"
IndianaDF = pd.read_csv(IndianaData)
IndianaDF
```

```python
#use of uszipcode starts here some modifications made to keep from reovewriting modified data
search = SearchEngine(simple_zipcode=False)
```

```python
Lat = []
Long = []
county = []
ZipCodes = IndianaDF["Zip"]
for littleZips in ZipCodes:
    zipcoding = search.by_zipcode(littleZips)
    ZipVar = zipcoding.to_dict()
    Lat.append(ZipVar["lat"])
    Long.append(ZipVar["lng"])
    county.append(ZipVar["county"])
Lat
```
