```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key

gmaps.configure(api_key=g_key)
mergedDATA = "merged_df.csv"
countyCoorData = "county coordinates.csv"
infantDeathData = "underlying_cause_final.csv"
indianaOBGYNData = "Indiana_OBGYNS_Final.csv"
infantDeathDF = pd.read_csv(infantDeathData)
indianaOBGYNDF = pd.read_csv(indianaOBGYNData)
countyCoor = pd.read_csv(countyCoorData)
mergedDF = pd.read_csv(mergedDATA)
indianaOBGYNDF
```

```python
coordinatesNA = indianaOBGYNDF[['Lat', 'Long']]
coordinates = coordinatesNA.dropna()
coordinates
```

```python
f = gmaps.figure()

heat_layer = gmaps.heatmap_layer(coordinates, dissipating=True, max_intensity=10)
f.add_layer(heat_layer)
f
```

```python
#infantDeathDF.loc[infantDeathDF['CHILD_DEATH_YR'] == '2018']
values = ['2018']
iDeath2018 = infantDeathDF.loc[infantDeathDF['CHILD_DEATH_YR'].isin(values)]
iDeath2018
```

```python
iDeath2018.columns = ['County', 'Death Year', 'Deaths per Year', 'Year Deaths per 10000 Residents']
coordinatesNAB = countyCoor[['County','Lat', 'Lng']]
coordinates2 = coordinatesNAB.dropna()
iDeath2018
```

```python
Merged1 = pd.merge(mergedDF, coordinates2, on='County', how='outer')
Merged2 = pd.merge(Merged1, iDeath2018, on='County', how='outer')
Merged3 = Merged2[np.isfinite(Merged2['Lat'])]
Merged3["Lng"] = -Merged3["Lng"]
MergedFill = Merged3.fillna(0)
#MergedFill.to_csv('MergedAgain.csv')
```

```python
MergedFill2Data = 'MergedAgain.csv'
MergedFill2 = pd.read_csv(MergedFill2Data)
Deaths = MergedFill2["Deaths per Year"]
coordinatesHeat = MergedFill2[['Lat2', 'Lng']]
fig = gmaps.figure()

heat_layer2 = gmaps.heatmap_layer(coordinatesHeat, weights=Deaths,dissipating=True, max_intensity=5)
fig.add_layer(heat_layer2)
fig
```

```python
DeathsPer = MergedFill2["Year Deaths per 10000 Residents"]
fig2 = gmaps.figure()

heat_layer3 = gmaps.heatmap_layer(coordinatesHeat, weights=DeathsPer,dissipating=True, max_intensity=1)
fig2.add_layer(heat_layer3)
fig2
```
